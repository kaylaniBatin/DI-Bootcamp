```python
import pandas as pd

# Sample data
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'London', 'Tokyo']
}

# Create DataFrame
df = pd.DataFrame(data)

# Display it
df

```

```python
# Install openpyxl for writing Excel .xlsx
!pip install openpyxl

# Export to Excel .xlsx
df.to_excel("data_export.xlsx", index=False, engine='openpyxl')

# Download the Excel file
from google.colab import files
files.download("data_export.xlsx")
```

```python
# Export to JSON
df.to_json("data_export.json", orient='records', indent=2)

# Download JSON
files.download("data_export.json")
```
